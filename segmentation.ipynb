```python
import warnings
warnings.simplefilter(action ='ignore', category =FutureWarning)
warnings.simplefilter(action ='ignore', category =Warning)



import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
loan_full_df=pd.read_csv('/content/accepted_2007_to_2018Q4.csv')

```

```python
loan_full_df.head()
```

```python
loan_full_df['loan_status'].value_counts()
```

```python
loan_int_df=loan_full_df[loan_full_df['loan_status']!='Current']
loan_int_df=loan_int_df[loan_full_df['loan_status']!='Late (31-120 days)']
loan_int_df=loan_int_df[loan_full_df['loan_status']!='In Grace Period']
loan_int_df=loan_int_df[loan_full_df['loan_status']!='Late (16-30 days)']
loan_int_df['loan_status']=loan_int_df['loan_status'].replace('Charged Off','Default')
loan_int_df['loan_status']=loan_int_df['loan_status'].replace('Fully Paid','Good')
```

```python
loan_int_df['loan_status'].value_counts()
```

```python
loan_int_df['last_fico_range_low'].min()
```

```python
loan_int_df['last_fico_range_low']=loan_int_df.apply(lambda x: x['fico_range_low'] if x['last_fico_range_low']==0.0 else x['last_fico_range_low'],axis=1)
```

```python
loan_int_df['last_fico_range_low'].value_counts()
```

```python
loan_int_df['last_fico_range_low'].value_counts().sort_index()
```

```python
pd.cut(loan_int_df['last_fico_range_low'],5,labels=['Very Low','Low','Medium','High','Very High'])
```

```python
loan_int_df['fico_bucket']= pd.cut(loan_int_df['last_fico_range_low'],5,labels=['Very Low','Low','Medium','High','Very High'])
```

```python
loan_int_df['fico_bucket'].value_counts()
```
